```python
# Import the necessary libraries
from dotenv import load_dotenv
from PyPDF2 import PdfReader
from langchain.text_splitter import CharacterTextSplitter
from langchain.embeddings.openai import OpenAIEmbeddings
from langchain.vectorstores import FAISS
from langchain.chains.question_answering import load_qa_chain
from langchain.llms import OpenAI
from langchain.callbacks import get_openai_callback
import os
from google.colab import files
```

```python
# Load environment variables from .env file
load_dotenv()
```

```python
def main():
    # Upload PDF file
    uploaded = files.upload()  # Use this for file upload in Colab

    if uploaded:
        file_name = next(iter(uploaded))  # Get the file name
        pdf_reader = PdfReader(file_name)
        text = ""
        for page in pdf_reader.pages:
            text += page.extract_text()

        # Split text into chunks
        text_splitter = CharacterTextSplitter(
            separator="\n",
            chunk_size=1000,
            chunk_overlap=200,
            length_function=len
        )
        chunks = text_splitter.split_text(text)

        # Create embeddings
        embeddings = OpenAIEmbeddings()
        knowledge_base = FAISS.from_texts(chunks, embeddings)

        # Get user question
        user_question = input("Ask a question about your PDF: ")

        if user_question:
            # Search for relevant chunks
            docs = knowledge_base.similarity_search(user_question)

            # Load the OpenAI LLM and QA chain
            llm = OpenAI()
            chain = load_qa_chain(llm, chain_type="stuff")

            # Run the chain and get response
            with get_openai_callback() as cb:
                response = chain.run(input_documents=docs, question=user_question)
                print(cb)

            # Display the response
            print("Answer:", response)

if __name__ == '__main__':
    main()
```
